# Landsat Imagery

```python
from arcgis.gis import GIS
```

```python
from arcgis.raster.functions import *
```

```python
gis = GIS()
```

```python
landsat_item = gis.content.search('Landsat Multispectral', 'Imagery Layer')[0]
```

```python
landsat_item
```

```python
landsat = landsat_item.layers[0]
```

# Raster Functions

```python
def extract_stretch(bandids):
    return stretch(extract_band(landsat, bandids),
                   stretch_type='PercentClip',
                   min_percent=2,
                   max_percent=2,
                   dra=True,
                   gamma=[0.8, 0.8, 0.8])
```

## Cambridge Gulf, Australia

```python
from arcgis.geocoding import geocode
area = geocode('Cambridge Gulf')[0]
```

```python
m = gis.map(area, 10)
m
```

```python
m.add_layer(extract_stretch([5, 4, 1]))
```

```python
cambridge_gulf = extract_stretch([5, 4, 1])
cambridge_gulf.extent = {'xmax': 128.96,
                         'xmin': 127.62,
                         'ymax': -14.72,
                         'ymin': -15.25,
                         'spatialReference': 4326}
cambridge_gulf
```
